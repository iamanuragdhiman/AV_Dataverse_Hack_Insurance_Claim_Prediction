## Imbalance Dataset Problem:
   
* Initially I tried to run various algorithms and try to get the classification metrics and accuracy on the dataset.
* But this dataset contains 58K rows and my system was not capable enough to run ensemble methods using gridsearch and I had to wait for hours to reach to the best parameters.
* I wasted a lot of time in that, so as per my understanding I went ahead with the resampling approach.
* Resampling (Oversampling and Undersampling):
    * This technique is used to upsample or downsample the minority or majority class. When we are using an imbalanced dataset, we can oversample the minority class using replacement. This technique is called oversampling. Similarly, we can randomly delete rows from the majority class to match them with the minority class which is called undersampling. After sampling the data we can get a balanced dataset for both majority and minority classes. So, when both classes have a similar number of records present in the dataset, we can assume that the classifier will give equal importance to both classes.
    * As it was not possible for me to oversample the data due to lack of computing resources, I was left with the only option to undersample the data.

## Word of caution:

* Undersampling is not a favourable technique, since we are trying to undersample the data and reduce the number of observations.
* This indeed means that we are losing information and we will not fit the data well.
* This will result in lesser accuracy but advantage is we can manage this prediction task on the low computing resource.
* Let's move ahead with the undersampling.

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42, replacement=True)# fit predictor and target variable
X_rus, y_rus = rus.fit_resample(X, y)

print('original dataset shape:', len(y))
print('Resample dataset shape', len(y_rus))
```

```python
X_rus.shape,y_rus.shape
```

```python
df_new = pd.concat([X_rus,y_rus],axis=1)
```

```python
df_new.head()
```

```python
df_new.is_claim.value_counts()
```

## After Undersampling:
* Undersampling has brought the number of observations drastically.
* We have now 7496 observations instead of 58K observations.
* But now we have balanced dataset.

```python
# sample 20% of data to be used as unseen data
data = df_new.sample(frac=0.8, random_state=42)
data_unseen = df_new.drop(data.index)

data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)

# print the revised shape
print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

## Why Pycaret?

* As I had lost plenty of time earlier, I did not have much time to run the manual process of validating multiple models.
* PyCaret is a wonderful auto ML library, we can compare and provide with the best results in quick steps and lesser time.
* So, here i have used this new approach, I am using this for the very first time.

```python
from pycaret.classification import *
dataset = setup(data = data, target = 'is_claim')
```

## Comparison of models in PyCaret:

* This is a balanced dataset now, we can use accuracy, F1 Score to evaluate the models.
* However the precision and recall are important in this dataset, so I have used F1-score as the evalation metric in this case.

```python
best_model = compare_models()
```

```python
gbc = create_model('gbc')
```

```python
rf = create_model('rf')
```

```python
tuned_gbc = tune_model(gbc,optimize='F1')
```

```python
print(tuned_gbc)
```

```python
tuned_rf = tune_model(rf,optimize='F1')
```
